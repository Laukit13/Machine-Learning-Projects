```python
## data Cleaning 
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Pclass',y='Age',data=data,palette='winter')
```

```python
def decnull_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 38

        elif Pclass == 2:
            return 29

        else:
            return 25

    else:
        return Age
```

```python
data['Age'] = data[['Age','Pclass']].apply(decnull_age,axis=1)
```

```python
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
data.drop('Cabin',axis=1,inplace=True)
```

```python
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
data.dropna(inplace=True)
```

```python
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
### Converting Categorical Features

### We'll need to convert categorical features to dummy variables using pandas! Otherwise our machine learning algorithm won't be able to directly take in those features as inputs.
```

```python
data.info()
```

```python
pd.get_dummies(data['Embarked'],drop_first=True).head()
```

```python
sex2 = pd.get_dummies(data['Sex'],drop_first=True)
embarked2 = pd.get_dummies(data['Embarked'],drop_first=True)
```

```python
data.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
data.head()
```

```python
data = pd.concat([data,sex2,embarked2],axis=1)
data.head()
```

```python
###Building a Logistic Regression model
###Let's start by splitting our data into a training set and test set (there is another test.csv file that you can play around with in case you want to use all this data for training).
```

```python
data.drop('Survived',axis=1).head()
```

```python
data['Survived'].head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop('Survived',axis=1),data['Survived'], test_size=0.30,random_state=101)
```

```python
### Traning and Predicting
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
accuracy=confusion_matrix(y_test,predictions)
```

```python
accuracy
```

```python
accuracy=accuracy_score(y_test,predictions) * 100
accuracy
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
