```python
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
#from sklearn.feature_extraction.text import tfvectorizer
from sklearn.decomposition import TruncatedSVD

post = open ('raw_forum_posts.dat','r')
#.read()

Soup = BeautifulSoup(post)
postTxt= Soup.find_all('text')
postDocs = [x.text for x in postTxt]
postDocs.pop(0)
postDocs = [x.lower() for x in postDocs]
```

```python
stopset = set(stopwords.words('english'))
stopset.update(['lt','p','/p','amp','quot','field','spacing','text','none','family','line',
                'none','sans','serif','transform','Opt','16','roman','letter'])
```

```python
postDocs[0]
```

```python
vectorizer = TfidfVectorizer(stop_words=stopset,use_idf=True, ngram_range=(1,3))
X = vectorizer.fit_transform(postDocs)
```

```python
X[0]
```

```python
print(X[0]) 
```

```python
X.shape
```

```python
lsa =TruncatedSVD(n_components=27,n_iter=100)
lsa.fit(X)
```

```python
lsa.components_[0]
```

```python
terms = vectorizer.get_feature_names()
for i, comp in enumerate(lsa.components_):
        termsIncomp = zip (terms , comp)
        sortedterms = sorted(termsIncomp, key=lambda x: x[1], reverse=True) [:10]
        print ("Concept :", i)
        for term in sortedterms:
                print(term[0])
                print ("")
```
